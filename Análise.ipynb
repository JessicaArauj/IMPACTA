### Gerando relatório automático em pdf

```python
from reportlab.lib.pagesizes import letter
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Paragraph, Image, Spacer
from reportlab.lib.styles import getSampleStyleSheet

# Estilos texto
styles = getSampleStyleSheet()
styles['Normal'].alignment = 4

# Análises em variaveis string
analise_turma = """
Durante a análise de avaliação de impacto sobre o aumento do faturamento após o programa por turma e correlacionando a quantidade de inscritos por turma, observou-se que o aumento de faturamento por turma está relacionado também à quantidade de inscritos por turma, onde os dados sugerem proporcionalidade.
"""
analise_empregabilidade = """
Ao analisar a empregabilidade após o programa, observou-se que apenas 30% dos participantes aumentaram o número de colaboradores em seu quadro. Partindo-se do pressuposto de que esse impacto pode levar ao aumento de mão-de-obra direta e indireta, percebe-se que não foi analisado o impacto gerado indiretamente, que pode ser considerado a partir do aumento de faturamento. E, também, o tempo necessário para a maturação do negócio após a participação do projeto Inovativa.
"""
analise_faturamento = """
A partir da análise do faturamento comparativo dos períodos pós-programa e anterior, percebe-se que 58% dos participantes declararam aumento superior a 50% no faturamento de seus negócios. Posteriormente, foi realizada a análise exploratória para entender o percentual geral dos participantes com faturamento positivo após o projeto, cerca de 94,9% do total de participantes declararam aumento em sua receita.
"""
analise_significancia = """
Para a realização abrangente das perguntas investigativas quanto à importância do impacto do programa gerado para os participantes, foi realizada a seguinte abordagem: soma das respostas para cada categoria em todas as colunas; cálculo dos percentuais obtidos em relação ao total de respostas; e geração de gráfico de setores. Após a iteração das perguntas, observou-se que, de forma geral, em relação à importância gerada para o beneficiário: 57,2% do total de respostas foram avaliadas como muito importante e 28,7% do total de respostas foram avaliadas como importante. Portanto, esses resultados indicam que o programa teve um impacto positivo em cerca de 85,9% dos beneficiários.
"""
# Criando uma lista para armazenar os elementos do relatório
conteudo = []
# Adicionando as analise para gráfico correspondente
conteudo.append(Paragraph("Análise do perfil:", styles['Normal']))
conteudo.append(Paragraph(analise_turma, styles['Normal']))
imagem_grafico1 = 'turma.png'
imagem_grafico2 = 'faturamentoturma.png'
conteudo.append(Image(imagem_grafico1, width=400, height=400))
conteudo.append(Image(imagem_grafico2, width=400, height=400))
conteudo.append(Spacer(1, 20))

conteudo.append(Paragraph("Análise de empregabilidade:", styles['Normal']))
conteudo.append(Paragraph(analise_empregabilidade, styles['Normal']))
imagem_grafico3 = 'empregosgerados.png'
conteudo.append(Image(imagem_grafico3, width=400, height=300))
conteudo.append(Spacer(1, 20))

conteudo.append(Paragraph("Análise do faturamento:", styles['Normal']))
conteudo.append(Paragraph(analise_faturamento, styles['Normal']))
imagem_grafico4 = 'faturamentocomparativo.png'
imagem_grafico5 = 'faturamento.png'
conteudo.append(Image(imagem_grafico4, width=400, height=300))
conteudo.append(Image(imagem_grafico5, width=400, height=300))
conteudo.append(Spacer(1, 20))

conteudo.append(Paragraph("Análise da importância percebida:", styles['Normal']))
conteudo.append(Paragraph(analise_significancia, styles['Normal']))
imagem_grafico6 = 'importancia.png'
conteudo.append(Image(imagem_grafico6, width=400, height=300))
conteudo.append(Spacer(1, 20))

# Nome do arquivo
pdf_filename = 'relatorio_exploratorio.pdf'

# Gerando arquivo e adicionando o conteúdo
pdf = SimpleDocTemplate(pdf_filename, pagesize=letter)
pdf.build(conteudo)
```

### Envio do relatorio exploratório de forma automatizada por e-mail:

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.application import MIMEApplication

# Função do envio de e-mail com o relatório em anexo
def enviar_email(destinatario, assunto, mensagem, anexo):
    # Configurações do servidor SMTP (Gmail)
    smtp_host = 'smtp.gmail.com'
    smtp_port = 587
    usuario_email = 'seu_email@gmail.com'  # Insira seu e-mail
    senha_email = 'sua_senha_do_email'  # Insira sua senha

    # Configuração do e-mail
    remetente = usuario_email
    destinatarios = [destinatario]
    
    # Criar a mensagem do e-mail
    msg = MIMEMultipart()
    msg['From'] = remetente
    msg['To'] = ', '.join(destinatarios)
    msg['Subject'] = assunto
    
    # Corpo da mensagem do e-mail
    msg.attach(MIMEText(mensagem, 'plain'))
    
    # Anexar o arquivo PDF do relatório ao e-mail
    with open(anexo, 'rb') as arquivo:
        anexo_part = MIMEApplication(arquivo.read())
        anexo_part.add_header('Content-Disposition', 'attachment', filename=anexo)
        msg.attach(anexo_part)

    # Conectar ao servidor SMTP e enviar o e-mail
    with smtplib.SMTP(smtp_host, smtp_port) as server:
        server.starttls()
        server.login(usuario_email, senha_email)
        server.sendmail(remetente, destinatarios, msg.as_string())

# Chamando a função do envio de e-mail com o relatório em anexo
destinatario_email = 'destinatario@gmail.com'  # Insira o e-mail do destinatário
assunto_email = 'Relatório Exploratório de Dados'
mensagem_email = 'Prezado destinatário,\n\nSegue em anexo o relatório exploratório de dados.\n\nAtenciosamente,\nSeu Nome'
arquivo_pdf = 'relatorio_exploratorio.pdf'  # Insira o nome do arquivo PDF gerado

enviar_email(destinatario_email, assunto_email, mensagem_email, arquivo_pdf)
```
